```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers,models
```

```python
print(tf.__version__)
```

```python
x = np.array([[0,0],[0,1],[1,0],[1,1]],dtype=float)
y= np.array([[0],[1],[1],[0]],dtype=float)
```

```python
model = models.Sequential(
    [
      layers.Input(shape=(2,)),
      layers.Dense(1,activation='sigmoid')
    ]
)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(x,y,epochs=1000,verbose=1)
```

```python
plt.plot(history.history['loss'],label='loss',color='blue')
plt.plot(history.history['accuracy'], label='Accuracy', color='orange')
plt.xlabel('Epochs')
plt.ylabel('Value')
plt.title('Training Accuracy and Loss')
plt.legend()
plt.show()
```

```python
# test=np.array([1,1])
# print(model.predict(np.expand_dims(test, axis=0)))
# model.predict(np.array([[1, 1]]))
model.predict(x)
print(f"Inputs\t\t  Predictions")
for inp,pred in zip(x,model.predict(x)):
  binary = 1 if pred>0.5 else 0
  print(f"{inp}:->\t{binary}")
```
